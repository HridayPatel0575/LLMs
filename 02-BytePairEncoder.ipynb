```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding('gpt2')
```

```python
text =("Hello World.<|endoftext|>this is my code  <endoftext> thisiscodebook")
integers = tokenizer.encode(text,allowed_special={'<|endoftext|>'})

print(integers)
```

```python
string = tokenizer.decode(integers)
print(string)
```

```python
integer = tokenizer.encode('laslddo ddjfj')
print(integer)

string = tokenizer.decode(integer)
print(string)
```
